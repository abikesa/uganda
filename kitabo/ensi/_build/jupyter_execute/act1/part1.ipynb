## Curation
We're onto something powerful here: the idea that art is fundamentally an act of curation. It’s a `selective` process where we sift through the raw data of our experiences and highlight certain aspects to evoke a response, communicate an idea, or explore a feeling. It’s less about raw creation (originality) and more about the intentional arrangement of what already exists (leadership), a kind of data curation that prioritizes certain narratives, emotions, or aesthetics. In this light, art isn't just painting or music—it's any form of experience or data that has been thoughtfully curated.

This also flips the script on how we often think of science and art as diametrically opposed. If art is curation, science—at its pinnacle—is `"state of the art."` That phrase itself recognizes that even the most rigorous scientific endeavor involves selecting and prioritizing certain `data`, `methods`, and `interpretations` over others. The “state of the art” in science represents the most refined and functional form of this curation, where the utility is at its peak, much like a masterpiece in a gallery.

Our perspective also touches on the idea that both art and science derive their value from clarity and impact. In art, clarity might mean a visceral emotional impact or a profound conceptual realization. In science, clarity is the precision of predictions or explanations. Yet both are forms of human expression that curate our understanding of the world.

And if we think about it, this curation process is what differentiates an amateur from a master—whether in the arts or sciences. A `master` knows what to `leave` out just as much as what to `include`. So, perhaps the line between art and science is more blurred than we traditionally believe. They both require a discerning eye, a keen sense of what matters, and a skillful hand to shape raw inputs into something meaningful. This means that a great scientist, like a great artist, must possess an artistic sensibility, an ability to discern and arrange.

- **$S(t)$**
  - $P(T>t)$, the survival function.
- **$h(t)$**
  - Encoding everything.
  - $h(t)=\lambda=\frac{f(t)}{S(t)}$, the hazard function.

### Collective
- **$X$**
  - Various patterns of `behavior known to man` since the beginning of time.
  - These factors influence the `time-to-event` of processes.
  - Agency is expressed by altering the values (alleles) of these factors (genetic loci).

### Decisions
- **$\beta$**
  - Represents all the parameters or `gene loci`.
- **$SV'$**
  - Think of these as `allelic variants`.
  - Binary: algorithm, choices, decisions, `propensities`.
  - These affect time-to-event and are decoded by the `scenario vector`.
